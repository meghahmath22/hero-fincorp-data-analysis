# Hero FinCorp Data-Driven Analysis

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (8,5)
```

```
# Load datasets (update file paths if needed)
customers = pd.read_csv('customers.csv')
loans = pd.read_csv('loans.csv')
applications = pd.read_csv('applications.csv')
transactions = pd.read_csv('transactions.csv')
defaults = pd.read_csv('defaults.csv')
branches = pd.read_csv('branches.csv')
```

## Data Cleaning

```
# Handle missing values
customers = customers.drop_duplicates()
loans = loans.drop_duplicates()

# Fill missing numeric values
for df in [customers, loans, applications, transactions, defaults, branches]:
    for col in df.select_dtypes(include=np.number).columns:
        df[col].fillna(df[col].median(), inplace=True)
```

## Feature Engineering

```
# Create Default Flag
defaults['Default_Flag'] = 1
loans = loans.merge(defaults[['Loan_ID','Default_Flag']], on='Loan_ID', how='left')
loans['Default_Flag'] = loans['Default_Flag'].fillna(0)

# EMI calculation (approx)
loans['EMI'] = loans['Loan_Amount'] * (1 + loans['Interest_Rate']/100) / loans['Loan_Term']
```

## Exploratory Data Analysis

```
sns.histplot(loans['Loan_Amount'], kde=True)
plt.title("Loan Amount Distribution")
plt.show()

sns.histplot(customers['Credit_Score'], kde=True)
plt.title("Credit Score Distribution")
plt.show()
```

## Default Risk Analysis

```
corr = loans[['Loan_Amount','Interest_Rate','EMI','Default_Flag']].corr()
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Customer Segmentation

```
customers['Segment'] = pd.qcut(customers['Credit_Score'], 3, labels=['Low','Medium','High'])
customers['Segment'].value_counts()
```
